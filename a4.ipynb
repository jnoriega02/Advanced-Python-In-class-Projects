# 1. Read Data (15 pts)

use an iterator to read the file food-prices-weekly.txt into a list of dictionaries named data

```python
def file_extract(filename):
    open_file=open(filename,'r',encoding='utf-16')
    clean_data=[]
    headings=[]
    # extracted_info=[extracted_info.append()]
    #get the headers as cat
    for pos,line_num in enumerate(open_file):
        if pos in [0]:
           headings= line_num.split() #split 
        else:
            entry=dict(zip(headings,[line_num[0:14].strip(),line_num[16:26],line_num[27:55].strip(),line_num[55:66].strip(),line_num[66:77].strip()])) #add tuples made from zip as dictionary 
            
            clean_data.append(entry)
            # print(entry)
            

    return (clean_data)             
   
file_extract('food-prices-weekly.txt') # call function
    
    
```

# 2. Add Month Column (5 pts)

For each data item, create a new key-value pair, Month, from the Date column. Given a date in yyyy-mm-dd format, the Month value should be yyyy-mm. Add these pairs to the existing dictionaries in the list.

```python
data_list=file_extract('food-prices-weekly.txt')

def add_month(list):
    
    for x in list:
        if (x.get('Date')): #get date
            placeholder=(x['Date'].split('-')) #split and pop day of date
            placeholder.pop()
            month='-'.join(placeholder)
        
            x["Month"]=month        
    return list
            
          
add_month(data_list)

            
        
```

# 3. Compute Monthly Values (15/25 pts)

Create a new list of dictionaries named monthly_data that has one entry per month for each state and category that contains four key-value pairs (State, Month, Category, DollarsAvg). DollarsAvg should be the average of the Dollars values for the given month, state, and category

```python

added_col_list=add_month(data_list)

def create_new_month_list(data_list):
    aggregated_data = {}
    monthly_list = []
    for entry in data_list:
        
        dollars = int(entry['Dollars']) #convert to int
        
        
        last_year = int(entry['LastYear']) if entry['LastYear'] != '-999' else None
        
        key = (entry['State'], entry['Month'], entry['Category'])
        
        #if not already a dict make it one
        if key not in aggregated_data:
            aggregated_data[key] = {'dollars_sum': 0, 'dollars_count': 0, 'last_year_sum': 0, 'last_year_count': 0}
        
        aggregated_data[key]['dollars_sum'] += dollars
        #calculate average
        aggregated_data[key]['dollars_count'] += 1
        
        #calcualte last year abg
        if last_year is not None:
            aggregated_data[key]['last_year_sum'] += last_year
            
            aggregated_data[key]['last_year_count'] += 1
    
    
    
    for (state, month, category), values in aggregated_data.items():
        
        
        dollars_avg = values['dollars_sum'] / values['dollars_count']
        if values['last_year_count'] > 0:
            
            last_year_avg = values['last_year_sum'] / values['last_year_count']
            pct_change = ((dollars_avg - last_year_avg) / last_year_avg) * 100 if last_year_avg else None
        else:
            last_year_avg = None
            pct_change = None
        #add to final list
        monthly_list.append({
            'Month': month,
            'State': state,
            'Category': category,
            'DollarsAvg': dollars_avg,
            'LastYearAvg': last_year_avg,
            'PctChange': pct_change
        })
    
    return monthly_list

            
            
create_new_month_list(added_col_list)         

```

# 4. Write Output (20/25 pts)

Write the new monthly_data list of dictionaries to a file named food-prices-monthly.txt in a the standard utf-8 format. This means writing the header values as the first line. Second, write the floating point values for DollarsAvg (and for CSCI 503, LastYearAvg) with only 1 digit after the decimal point. CSCI 503 students should also write PctChange with 2 digits after the decimal point. All the numbers should be right-aligned with at least two spaces between each column. Write a positive sign (+) for positive values of PctChange and a negative sign for negative values. The positions of each entry (inclusive) should be:

```python
# Define the header and the format function again for clarity
header = "State           Month    Category                      DollarsAvg  LastYearAvg  PctChange"
monthly_data=create_new_month_list(added_col_list)         

def format_entry(entry):
    # Updated to handle None values gracefully for LastYearAvg and PctChange
    dollars_avg = f"{entry['DollarsAvg']:10.1f}"
    
    last_year_avg = f"{entry['LastYearAvg']:10.1f}" if entry['LastYearAvg'] is not None else "         -"
    
    pct_change = f"{entry['PctChange']:>+7.2f}" if entry['PctChange'] is not None else "      -"
    return f"{entry['State']:14}{entry['Month']:7}  {entry['Category']:26}{dollars_avg}{last_year_avg}{pct_change}"



# Writing to the file with a relative path


with open('food-prices-monthly.txt', 'w', encoding='utf-8') as file:
    print(header, file=file)  # Write the header
    
    for entry in monthly_data:
        print(format_entry(entry), file=file)  # Write each formatted entry


#works in jupyter not vs
 !cat food-prices-monthly.txt
```
