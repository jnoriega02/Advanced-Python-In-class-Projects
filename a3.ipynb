# 4. Sales Sums (15 pts)

Write code to create a dictionary that keeps track of the sum of the sales that each senator has made.

```python
def sale_sums(trans_list):
    my_dict={}
    for x in trans_list:
        if('Sale' in x.get('type')):
            senator_key=next(reversed(x))
            senator_name=x[senator_key]
            
            initial_range=x.get('amount_range',(0,0))

            if senator_name not in my_dict: #set an inital value for first entry 
                my_dict[senator_name] = (0, 0) 

            current_sum = my_dict[senator_name]   #initial amount already in dict
            my_dict[senator_name]=(current_sum[0]+initial_range[0],current_sum[1]+initial_range[1]) #proper addition of ranges
    print('Sum of Sales made by Senators')        
    return my_dict

sale_sums(data)
```

# 5. Median Transaction by Ticker Symbol (15 pts)
Write a function get_ticker_median that, given a ticker symbol, returns the median transaction range for that ticker symbol. 

```python
def get_ticker_median(ticker,transaction_list):
    ticker_list=[] #empty list
    for x in transaction_list:
        if (x.get('ticker')==ticker):
            ticker_list.append(x.get('amount_range'))
            
            help_sort=lambda a: a[1] #lambda function to help  sort 
    ticker_list.sort(key=help_sort)
    
    length_list=(len(ticker_list)) 
    
    if (length_list%2>0):  #if list lengh is even
        while (len(ticker_list)>1):
            ticker_list.pop(-1)
            ticker_list.pop(0)
            print('The medium for the ticker:',ticker)
            return ticker_list
    else:
        while (len(ticker_list)>2):
            ticker_list.pop(-1)
            ticker_list.pop(0)
        first_range = ticker_list[0]  #GRABS THE FIRST FORM FIRST AND SECOND FROM SECOND RANGE
        second_range=ticker_list[1] 
        final_range=[first_range[0],second_range[1]]
        print('The medium for the ticker:',ticker)
        return final_range

get_ticker_median('AAPL',data)


            

    
            
```

# [CSCI 503 Only] Filter by Date (20 pts)

Write a function transactions_in_range that will filter the sales by date (inclusive). Specifically, given a start date and an end date, return the transactions that fall in that range. 

```python
def transactions_in_range(start, end, transaction_list):
    final_list = []
    start_date = list(map(int, start.split('/')))
    end_date = list(map(int, end.split('/')))
    
    for transaction in transaction_list:
        transaction_date = list(map(int, transaction.get('transaction_date').split('/')))
        if start_date <= transaction_date <= end_date:
            if('Sale' in transaction.get('type')):
                final_list.append(transaction)
            
    return final_list
transactions_in_range("1/17/2023", "1/20/2023",data)

              
```
